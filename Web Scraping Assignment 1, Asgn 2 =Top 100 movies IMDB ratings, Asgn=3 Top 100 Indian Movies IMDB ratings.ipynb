```python
# a python program to display all the header tags from‘en.wikipedia.org/wiki/Main_Page’

from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup
```

```python
html = urlopen('http://en.wikipedia.org/wiki/Main_Page')
```

```python
bs = BeautifulSoup(html,'html.parser')
```

```python
titles = bs.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
```

```python
print('List all the header tags :', *titles, sep='\n\n')
```

```python
# python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.imdb.com/chart/top/'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup =BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
scraped_movies = soup.find_all('td', class_='titleColumn')
scraped_movies
```

```python
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n','')
    movie = movie.strip(" ")
    movies.append(movie)
movies      
```

```python
scraped_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
scraped_ratings
```

```python
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n','')
    ratings.append(rating)
ratings    
```

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['Ratings'] =ratings
data.head(100)
```
