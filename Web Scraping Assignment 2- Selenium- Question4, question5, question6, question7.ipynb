```python
# question 5- : Scrape 100 reviews data from flipkart.com for iphone11 phone.

import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"C:\Users\lenovo\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url = "https://www.flipkart.com/apple-iphone-11-black-64-gb-includesearpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace"
```

```python
driver.get(url)
```

```python
view_all=driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']//span")
view_all.click()
```

```python
page_url=[]
url=driver.find_elements_by_xpath("//nav[@class='yFHi8N']//a")
for i in url:
    page_url.append(i.get_attribute("href"))
page_url
```

```python
iphone_rat=[] 
for i in page_url:
    driver.get(i)
    rat_tags=driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']")
    for i in rat_tags:
        iphone_rat.append(i.text)
iphone_rat[:100]
```

```python
review_sum=[] 
for i in page_url:
    driver.get(i)
    rev_tags=driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
    for i in rev_tags:
        review_sum.append(i.text)
review_sum[:100]
```

```python
full_rev=[] 
for i in page_url:
    driver.get(i)
    fure_tags=driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
    for i in fure_tags:
        full_rev.append(i.text)
full_rev[:100]
```

```python
import pandas as pd
iphone_11=pd.DataFrame({})
iphone_11['Ratings']=iphone_rat[:100]
iphone_11['Review summary']=review_sum[:100]
iphone_11['Full review']=full_rev[:100]
```

```python
#question 6- Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field. You have to scrape 4 attributes of each sneaker: Brand ,Product Description ,Price

import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"C:\Users\lenovo\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url='https://www.flipkart.com/'
driver.get(url)
```

```python
search_item=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_item
```

```python
search_item.send_keys('Sneakers')
```

```python
search_btn = driver.find_element_by_class_name('L0Z3Pu')

search_btn.click()
```

```python
page_url=[] 
url=driver.find_elements_by_xpath("//nav[@class='yFHi8N']//a")
for i in url:
    page_url.append(i.get_attribute("href"))
page_url
```

```python
snea_brands=[] 
for i in page_url:
    driver.get(i)
    brand_tags=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    for i in brand_tags:
        snea_brands.append(i.text)
snea_brands[:100]
```

```python
pro_description=[] 
for i in page_url:
    driver.get(i)
    des_tags=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
    for i in des_tags:
        pro_description.append(i.text)
pro_description[:100]
```

```python
price=[] 
for i in page_url:
    driver.get(i)
    pri_tags=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in pri_tags:
        price.append(i.text)
price[:100]
```

```python
discount=[] 
for i in page_url:
    driver.get(i)
    dis_tags=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']//span")
    for i in dis_tags:
        discount.append(i.text)
discount[:100]
```

```python
sneakers=pd.DataFrame({})
sneakers['Brand']=snea_brands[:100]
sneakers['Product description']=pro_description[:100]
sneakers['Price']=price[:100]
sneakers['Discount']=discount[:100]
```

```python
sneakers
```

```python
# Question 7- Myntra Shoes
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"C:\Users\lenovo\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url='https://www.myntra.com/shoes'
driver.get(url)
```

```python
price_filter=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
price_filter.click()
```

```python
color_filter=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
color_filter.click()
```

```python
br=driver.find_elements_by_xpath("//h3[@class='product-brand']")
```

```python
brand=[] 
for i in range(1,4):
    for j in br:
        brand.append(j.text)
    next_button=driver.find_element_by_xpath("//li[@class='pagination-next']/a")
    next_button.click
```

```python
len(brand)
```

```python
brand[:100]
```

```python
des=driver.find_elements_by_xpath("//h4[@class='product-product']")
```

```python
description=[] 
for i in range(1,4):
    for j in des:
        description.append(j.text)
    next_button=driver.find_element_by_xpath("//li[@class='pagination-next']/a")
    next_button.click
```

```python
len(description)
```

```python
description[:100]
```

```python
pri_tags=driver.find_elements_by_xpath("//div[@class='product-price']")
```

```python
price=[] 
for i in range(1,4):
    for j in pri_tags:
        price.append(j.text)
    next_button=driver.find_element_by_xpath("//li[@class='pagination-next']/a")
    next_button.click
```

```python
len(price)
```

```python
price[:100]
```
